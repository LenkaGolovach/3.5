Рассмотрим систему падающих налетающих объектов. Пусть воздушный поток налетающих объектов состоит из двух типов: сферических частиц и плоских пластинок. Сферические частицы имеют массу 0.1 кг и начальную скорость 10 м/с. Плоские пластинки имеют массу 0.05 кг и начальную скорость 15 м/с.

Необходимо построить график зависимости высоты объектов от времени и определить время, через которое каждый объект достигнет земли.

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

Задаем начальные условия для сферических частиц

```python
mass_sphere = 0.1  # Масса сферической частицы в кг
initial_velocity_sphere = 10  # Начальная скорость сферической частицы в м/с
```

Задаем начальные условия для плоских пластинок

```python
mass_plate = 0.05  # Масса плоской пластинки в кг
initial_velocity_plate = 15  # Начальная скорость плоской пластинки в м/с
```

Вычисляем время падения для сферических частиц

```python
time_sphere = np.linspace(0, 2, 100)  # Время от 0 до 2 секунды с равными интервалами
height_sphere = initial_velocity_sphere * time_sphere - 0.5 * 9.8 * time_sphere**2
```

Вычисляем время падения для плоских пластинок

```python
time_plate = np.linspace(0, 2, 100)  # Время от 0 до 2 секунды с равными интервалами
height_plate = initial_velocity_plate * time_plate - 0.5 * 9.8 * time_plate**2
```

```python
# Cтроим график
plt.plot(time_sphere, height_sphere, label='Сферические частицы')
plt.plot(time_plate, height_plate, label='Плоские пластинки')
plt.xlabel('Время, сек')
plt.ylabel('Высота, м')
plt.title('Движение налетающих объектов')
plt.legend()
plt.grid(True)
plt.show()
```
